```python
# Parete Chart
# libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# data frame
df = pd.DataFrame({"grasses": [92, 83, 76, 59, 53, 27, 16, 9, 7, 4, 3, 1]})

# Reset the indexes
df.index = ["LongDrinks", "FortifiedWine", "FlavoredWine", "ScotchWhisky", "SparklingWine", "Cocktails", "BrownSpirits", "NonAlcohlCocktails", "OtherSpririts", "StillWine", "Beer", "SoftDrinks"]

# Sort values: descending
df = df.sort_values(by='grasses', ascending=False)

# Add column: cumulative percentage
df["cum_percentage"] = round(df["grasses"].cumsum()/df["grasses"].sum()*100,2)
```

```python
df
```

```python
# fig axis
fig, ax = plt.subplots(figsize=(22,10))

# Plot:
# bar
ax.bar(df.index, df["grasses"])
ax.set_title("Pareto Chart")
ax.set_xlabel("Medication Grasses")
ax.set_ylabel("Frequency");

# Second y: cumulative percentage
ax2 = ax.twinx()
ax2.plot(df.index, df["cum_percentage"], color="red", marker="D", ms=7)
ax2.axhline(80, color="orange", linestyle="dashed")
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_ylabel("Cumulative Percentage");
```
